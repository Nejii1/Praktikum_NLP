```python
import re
import pandas as pd

# Baca teks dari file
with open('dummy_data.txt', 'r') as file:
    text = file.read()

# regex processing
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
emails = re.findall(email_pattern, text)

phone_pattern = r'(?:(?:\+?(\d{1,3}))?[-. (]?)?(\d{3})[-. )]?(\d{3})[-. ]?(\d{4})(?:x(\d+))?'
phone_numbers = re.findall(phone_pattern, text)

min_length = min(len(emails), len(phone_numbers))
emails = emails[:min_length]
phone_numbers = phone_numbers[:min_length]

# Gabung email dan nomor telepon
data = {'Nama': [], 'Nomor Telepon': [], 'Email': []}
current_name = ""

for line in text.split('\n'):
    if line.startswith("Nama: "):
        current_name = line.replace("Nama: ", "")
    elif line.startswith("Nomor Telepon: "):
        data['Nama'].append(current_name)
        data['Nomor Telepon'].append(line.replace("Nomor Telepon: ", ""))
    elif line.startswith("Email: "):
        data['Email'].append(line.replace("Email: ", ""))

df = pd.DataFrame(data)

# nampilin data frame
df
```
