# Soal 2

<br>Lakukan pembersihan meliputi penghapusan punctuation, angka dan karakter yang tidak penting menggunakan menggunakan regex.
<br>Simpan hasilnya menjadi file csv/excel.


```python
# Membersihkan kolom full_text
def clean_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

df_eval['cleaned_text'] = df_eval['full_text'].apply(clean_text)

df_eval
```

```python

def clean_username(username):
    cleaned_username = re.sub(r'[^a-zA-Z\s]', '', username)
    return cleaned_username

df_eval['cleaned_username'] = df_eval['username'].apply(clean_username)

df_eval
```

```python
df_clean = df_eval.drop(['full_text', 'username'], axis = 1)

df_clean
```

```python
df_clean.to_csv('data_cleaned.csv', index=False)
```

# Soal 3b

b.Carilah kata-kata slangword yang ada dalam dataset Anda, dengan cara mencocokan dengan kamus KBBI (terlampir).
<br>Simpan hasilnya dalam bentuk csv/excel.
<br> Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam bentuk Gunakan (bobot:25)

```python
# Membaca file CSV yang berisi kata-kata slang
slang_df = pd.read_csv('stopwords.csv')

slang_df
```

```python
slang_words = slang_df['Slangwords'].tolist()

# Fungsi untuk mencari kata-kata slang dalam teks
def find_slang(text):
    words = text
    found_slang = [word for word in words if word in slang_words]
    return ', '.join(found_slang)

# Menerapkan fungsi ke kolom 'parse_text'
data_parse['found_slang_in_text'] = data_parse['parse_text'].apply(find_slang)

data_parse
```

```python
# Mengambil kata-kata slang yang telah ditemukan
found_slang_list = data_parse['found_slang_in_text'].str.split(', ').explode().tolist()

# Mengambil 100 kata pertama
seratus_slang = found_slang_list[:100]

# Menampilkan 100 kata pertama
seratus_slang
```

# Soal 3d

d.Lakukan stopword removing  berdasarkan hasil 3c, simpan hasilnya dalam bentuk csv/excel. (bobot:10)!

```python
import nltk
nltk.download('stopwords')
```

```python

# Daftar kata-kata yang telah Anda berikan
daftar_kata = hasil_token

# Menggunakan NLTK untuk mendapatkan daftar stopword dalam bahasa Indonesia
stopwords_list = set(stopwords.words('indonesian'))

# Menghapus stopword dari daftar kata-kata
filtered_words = [word for word in daftar_kata if word not in stopwords_list]

# Menampilkan hasil setelah penghapusan stopword
filtered_words
```

```python
# Membuat DataFrame dari hasil penghapusan stopword
df_stopword = pd.DataFrame({'hasil_stopwords': filtered_words})

# Menyimpan hasilnya ke dalam file CSV
df_stopword.to_csv('hasil_stopwords.csv', index=False)
```

# Soal 3e

e.Lakukan stemming berdasarkan hasil 3d dan tampilkan 100 stem pertama. (bobot:10)

```python

# Daftar kata-kata setelah penghapusan stopword
word_list = pd.read_csv('hasil_stopwords.csv')

# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Melakukan stemming pada setiap kata dalam daftar
stemmed_words = [stemmer.stem(kata) for kata in word_list]

# Menampilkan hasil stemming
print(stemmed_words)
```
