# Proses TF-IDF

```python
teks_gabung = [' '.join(kalimat) for kalimat in teks_split]
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(teks_gabung)
```

```python
tfidf_matrix
```

```python
print(tfidf_matrix)
```

```python
vocab = tfidf.get_feature_names_out()
vocab
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=teks_gabung)
```

```python
dokumen_ke = 1 # Indeks data yang ingin diperiksa

indeks_kata_tertinggi = tfidf_matrix[dokumen_ke].argmax()

kata_tertinggi = vocab[indeks_kata_tertinggi]

bobot_tertinggi = tfidf_matrix[dokumen_ke, indeks_kata_tertinggi]

print(f"Kata dengan bobot tertinggi dalam dokumen {dokumen_ke} adalah '{kata_tertinggi}' dengan bobot {bobot_tertinggi:.2f}")
```
